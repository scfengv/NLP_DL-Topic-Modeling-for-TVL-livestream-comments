```python
import emoji
import warnings
import pandas as pd
```

```python
warnings.simplefilter("ignore")
```

```python
emoji_list = list(emoji.EMOJI_DATA.keys())
```

```python
def extract_rows(text):
    return any(char in emoji_list for char in text)
```

```python
df = pd.read_csv('./data/Overall_Layer_topic.csv')
result = pd.read_csv('./data/final_result.csv')
```

```python
df = df[['time', 'text', 'ÊØîË≥Ω', 'Âä†Ê≤π', 'ËΩâÊí≠', 'ÈñíËÅä']]
result = result[['time', 'text', 'ÁêÉÂì°', 'ÊïôÁ∑¥', 'Ë£ÅÂà§','Êà∞Ë°ì', 'ÁêÉÈöä', 'Âä†Ê≤π', 'ËΩâÊí≠', 'ÈñíËÅä']]
```

```python
result['ÊØîË≥Ω'] = result[['ÁêÉÂì°', 'ÊïôÁ∑¥', 'Ë£ÅÂà§', 'Êà∞Ë°ì', 'ÁêÉÈöä']].any(axis = 1).astype(int)
```

```python
result = result[['time', 'text', 'ÊØîË≥Ω', 'Âä†Ê≤π', 'ËΩâÊí≠', 'ÈñíËÅä']]
```

```python
merged_df = pd.merge(df, result, on = ['time', 'text'], how = 'inner', suffixes=('_True', '_Pred'))
```

```python
merged_df
```

```python
error = merged_df[(merged_df['ÊØîË≥Ω_True'] != merged_df['ÊØîË≥Ω_Pred']) | (merged_df['Âä†Ê≤π_True'] != merged_df['Âä†Ê≤π_Pred']) | (merged_df['ËΩâÊí≠_True'] != merged_df['ËΩâÊí≠_Pred']) | (merged_df['ÈñíËÅä_True'] != merged_df['ÈñíËÅä_Pred'])]
```

```python
error['text'] = error['text'].astype(str)
mask_error = error['text'].apply(extract_rows)
num_emoji_error = mask_error.sum()

result['text'] = result['text'].astype(str)
mask_all = result['text'].apply(extract_rows)
num_emoji_all = mask_all.sum()

print(f'Percentage of comments contains emoji: {round(num_emoji_all / len(result) * 100, 4)} %')
print(f'Percentage of error contains emoji: {round(num_emoji_error / len(error) * 100, 4)} %')
```
