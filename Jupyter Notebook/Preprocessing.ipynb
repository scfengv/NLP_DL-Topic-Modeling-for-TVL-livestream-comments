```python
emoji_list = emoji.EMOJI_DATA
```

```python
def extract_rows(text):
    return any(char in emoji_list for char in text)
    #  and all(char not in target_emoji for char in text)
```

```python
def emoji2description(text):
  return emoji.replace_emoji(text, replace = lambda chars, data_dict: ' ' + ' '.join(data_dict['zh'].split('_')).strip(':') + ' ')

def emoji2description_en(text):
  return emoji.replace_emoji(text, replace = lambda chars, data_dict: ' ' + ' '.join(data_dict['en'].split('_')).strip(':') + ' ')
```

```python
# def remove_punctuation_regex(input_string):
#     return re.sub(r'[^\w\s]', '', input_string)

def rmemoji(text):
  return emoji.replace_emoji(text, '').strip()
```

```python
chat_df_w_file_usr_time_text = pd.read_csv('./chat_df_w_file_usr_time_text.csv')
```

```python
df_emoji = chat_df_w_file_usr_time_text.copy()
```

```python
print(len(df_emoji))
df_emoji['text'] = df_emoji['text'].str.lower()
df_emoji['text'] = df_emoji['text'].str.replace(r'(:[^:]+:)', '')
df_emoji = df_emoji[(df_emoji['text'] != '') & (df_emoji['text'].notnull())]
print(len(df_emoji))
```

```python
df_remove_emoji = df_emoji.copy()
df_remove_emoji['text'] = df_emoji['text'].astype(str).apply(rmemoji)

print(len(df_remove_emoji))
df_remove_emoji = df_remove_emoji[(df_remove_emoji['text'] != '') & (df_remove_emoji['text'].notnull())]
print(len(df_remove_emoji))
```

```python
df_emoji_to_desc = df_emoji.copy()
df_emoji_to_desc['text'] = df_emoji['text'].astype(str).apply(emoji2description)

print(len(df_emoji_to_desc))
df_emoji_to_desc = df_emoji_to_desc[(df_emoji_to_desc['text'] != '') & (df_emoji_to_desc['text'].notnull())]
print(len(df_emoji_to_desc))
```

```python
df_emoji.to_csv('df_emoji.csv', index = False)
df_remove_emoji.to_csv('df_remove_emoji.csv', index = False)
df_emoji_to_desc.to_csv('df_emoji2desc.csv', index = False)
```
